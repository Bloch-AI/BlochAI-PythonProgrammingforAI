```python
import pandas as pd
import plotly.express as px

# Load the dataset (adjust the file path if necessary)
df = pd.read_csv("FIFA DATA.csv")

# ----------------------------------
# 1. Histogram: Distribution of Player Overall Ratings
# ----------------------------------
fig1 = px.histogram(
    df,
    x='OverallRating',  # Using 'OverallRating' column for player ratings
    nbins=20,
    title="Distribution of Player Overall Ratings",
    labels={"OverallRating": "Overall Rating", "count": "Number of Players"}
)
fig1.update_layout(bargap=0.1)
fig1.show()

# ----------------------------------
# 2. Scatter Plot: Age vs Overall Rating
# (Bubble size represents Player Value and colours represent Nationality)
# ----------------------------------
fig2 = px.scatter(
    df,
    x='Age',
    y='OverallRating',  # Using 'OverallRating' column for player ratings
    size='Value ',      # Using 'Value ' column to represent player value
    color='Nationality',
    hover_name='Name',  # Displays player's name on hover
    title="Player Age vs Overall Rating (Bubble size represents Value)",
    size_max=30         # Maximum bubble size for clarity
)
fig2.update_layout(xaxis_title="Age", yaxis_title="Overall Rating")
fig2.show()

# ----------------------------------
# 3. Bar Chart: Top 10 Clubs by Average Overall Rating
# ----------------------------------
# Calculate the average overall rating per club and sort in descending order
club_avg = df.groupby('Club', as_index=False)['OverallRating'].mean()
club_avg = club_avg.sort_values(by='OverallRating', ascending=False).head(10)

fig3 = px.bar(
    club_avg,
    x='Club',
    y='OverallRating',
    title="Top 10 Clubs by Average Overall Rating",
    labels={"OverallRating": "Average Overall Rating", "Club": "Club"}
)
fig3.update_layout(xaxis_tickangle=-45)
fig3.show()

# ----------------------------------
# Discussion: Power BI vs Python (Plotly) for Dashboarding
# ----------------------------------
# Advantages of using Python (Plotly):
# - High customisability and reproducibility: A code-based approach allows for detailed customisation of visualisations.
# - Integration: Easily integrates with other data science libraries for advanced analysis.
# - Automation: Dashboards and reports can be generated programmematically.
#
# Disadvantages:
# - Steeper learning curve: Requires programming knowledge.
# - Development time: Building interactive dashboards from scratch can be more time-consuming.
#
# Advantages of Power BI:
# - User-friendly interface: Drag-and-drop functionality makes it easier for non-programmers.
# - Quick prototyping: Faster to set up dashboards with built-in visuals and integrations.
#
# Disadvantages:
# - Limited customisation: Less flexibility compared to coding your own visualisations.
# - Dependency on proprietary software: Licensing and cost can be considerations.
```
