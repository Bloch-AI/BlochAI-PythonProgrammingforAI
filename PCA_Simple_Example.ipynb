```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Step 1: Create the dataset
data = pd.DataFrame({
    "Math": [85, 78, 92, 70, 88],
    "Science": [90, 85, 95, 75, 92],
    "History": [75, 80, 88, 65, 80]
}, index=["Alice", "Bob", "Charlie", "Dave", "Eve"])

# Step 2: Standardise the data (PCA works best when data is scaled)
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

# Step 3: Apply PCA
pca = PCA(n_components=2)  # Reduce to 2 components
principal_components = pca.fit_transform(scaled_data)

# Step 4: Convert PCA output to a DataFrame
pca_df = pd.DataFrame(principal_components, columns=["PC1 (Overall Score)", "PC2 (Science vs History)"], index=data.index)

# Step 5: Plot the PCA results
plt.figure(figsize=(8, 6))
plt.scatter(pca_df["PC1 (Overall Score)"], pca_df["PC2 (Science vs History)"], color=['red', 'blue', 'green', 'purple', 'orange'])

# Add labels to points
for i, name in enumerate(pca_df.index):
    plt.text(pca_df.iloc[i, 0], pca_df.iloc[i, 1], name, fontsize=12)

plt.axhline(0, color='black', linewidth=0.5, linestyle="--")
plt.axvline(0, color='black', linewidth=0.5, linestyle="--")
plt.xlabel("PC1 (Overall Score)")
plt.ylabel("PC2 (Science vs History)")
plt.title("PCA Projection of Student Scores")
plt.show()
```
